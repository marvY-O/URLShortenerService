```python
import random
import string

# Create an instance of the Faker class
fake = Faker()
urls = []

# Generate synthetic URLs
num_urls = 20000  # Adjust the number of URLs you want to generate
synthetic_urls = [fake.url() for _ in range(num_urls)]

def generate_random_string(min_length=80, max_length=120):
    length = random.randint(min_length, max_length)
    characters = string.ascii_letters + string.digits + '-_'
    return ''.join(random.choice(characters) for _ in range(length))

for url in synthetic_urls:
    urls.append(url+generate_random_string())
```

```python
df = pd.DataFrame({'url': urls})
```
